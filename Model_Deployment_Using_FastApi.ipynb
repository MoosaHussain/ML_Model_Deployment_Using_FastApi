```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
data = pd.read_csv("/content/drive/MyDrive/Bank_Note_Detection/BankNote_Authentication.csv")
data.head()
```

```
x = data[['variance','skewness','curtosis','entropy']]
y = data[['class']]
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3,train_size=0.7, random_state=42)
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score

clf = RandomForestClassifier()
clf.fit(X_train, y_train) 
y_pred=clf.predict(X_test)
acc=accuracy_score(y_test,y_pred)
acc
```

```
import pickle
pickle_out = open("classifier.pkl","wb")
pickle.dump(clf, pickle_out)
pickle_out.close()
```

```

import uvicorn
from fastapi import FastAPI
from Banknotes import BankNote
import numpy as np
import pickle
import pandas as pd
import nest_asyncio
from pyngrok import ngrok
# 2. Create the app object
app = FastAPI()
pickle_in = open("/content/drive/MyDrive/Bank_Note_Detection/classifier.pkl","rb")
classifier=pickle.load(pickle_in)

@app.get('/')
def index():
    return {'message': 'Hello, World'}

@app.get('/name')
def get_name(name: str):
  return {'Welcome To Bank Note Detection Model'}
 #   return {'Welcome To Bank Note Detection Model': f'{name}'}


#    JSON data and return the predicted Bank Note with the confidence
@app.post('/predict')
def predict_banknote(data:BankNote):
    data = data.dict()
    variance=data['variance']
    skewness=data['skewness']
    curtosis=data['curtosis']
    entropy=data['entropy']
   # print(classifier.predict([[variance,skewness,curtosis,entropy]]))
    prediction = classifier.predict([[variance,skewness,curtosis,entropy]])
    if(prediction[0]>0.5):
        prediction="Fake note"
    else:
        prediction="Its a Bank note"
    return {
        'prediction': prediction
    }
ngrok_tunnel = ngrok.connect(8000)
print('Public URL:', ngrok_tunnel.public_url)
nest_asyncio.apply()
uvicorn.run(app, port=8000)

    

```
